```python
from tkinter import *
import pymysql
import tkinter.messagebox as me
    
root = Tk()
root.title ('Entry window')
root.geometry("600x300")
root.configure(bg= 'Sky blue')

def createconn():
    return pymysql.connect(host='localhost',database='yashdb',
                           user='root',password='2580',port=3306)

def insertdata():
    r=ern.get()
    f=efn.get()
    l=eln.get()
    em=eem.get()
    mo=emo.get()
    
    if (r=="" or f=="" or l=="" or em=="" or mo==""):
        me.showinfo('Insert status','All files are mandatory')
    else:
        try:
            conn=createconn()
            cursor=conn.cursor()
            args=(r,f,l,em,mo)
            query= "insert into student(Sid,Fname,Lname,Email,Mobile) values (%s,%s,%s,%s,%s)"
            cursor.execute(query,args)
            conn.commit()
            me.showinfo('Insert status','Data Inserted')
            conn.close()
        except Exception as ee:
            print('Insert Exception: ',ee)
            
    
rn=Label(root,text='Roll No.')
rn.place(x=200,y=40)

fn=Label(root,text='First name')
fn.place(x=200,y=80)

ln=Label(root,text='Last name')
ln.place(x=200,y=120)

em=Label(root,text='Email ID')
em.place(x=200,y=160)

mo=Label(root,text='Mobile no.')
mo.place(x=200,y=200)

ern=Entry()
ern.place(x=300,y=40)

efn=Entry()
efn.place(x=300,y=80)

eln=Entry()
eln.place(x=300,y=120)

eem=Entry()
eem.place(x=300,y=160)

emo=Entry()
emo.place(x=300,y=200)


button1 = Button(root,text='Save',bg='Green',command=insertdata)
button1.place(x=200, y=250)



mainloop()
```
